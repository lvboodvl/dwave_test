# show computation result and draw figure of comparation of QA vs SA

```python
import numpy as np
from matplotlib import pyplot as plt
```

N:number of fiber span

```python
N = [100, 1000, 2000, 4000, 8000]
```

time of SA and QA

```python
time_SA_100 = [23.58, 56.75, 67.71, 84.93, 143.75]
time_SA_20 = [4.7, 11.08, 12.32, 17.05, 28.98]
time_QA_100 = [0.074, 0.139, 0.174, 0.245, 0.514]
time_QA_20 = [0.037, 0.086, 0.155, 0.211, 0.36]
```

|Sn| of SA and QA

```python
Sn_SA_100 = [0.253, 0.337, 2.432, 6.450, 15.932]
Sn_SA_20 = [3.038, 6.178, 11.89, 16.40, 26.714]
Sn_QA_100 = [0.01, 0.041, 0.16, 0.559, 1.567]
Sn_QA_20 = [0.057, 0.066, 0.407, 1.293, 2.356]
```

plot of |Sn|

```python
#plt.figure(figsize=(4, 4))
plt.rcParams['figure.figsize'] = (7.0, 7.0)

plt1 = plt.plot(N, Sn_QA_100, marker='o', mec='blue', mfc='w')
plt2 = plt.plot(N, Sn_QA_20, marker='*', mec='red', mfc='w')
plt3 = plt.plot(N, Sn_SA_100, marker='v', mec='black', mfc='w')
plt4 = plt.plot(N, Sn_SA_20, marker='s', mec='brown', mfc='w')
for a, b in zip(N, Sn_SA_20):
    plt.text(a, b+0.05, '%.02f' % b, ha='center', va='bottom', fontsize=14)
for a, b in zip(N, Sn_SA_100):
    plt.text(a, b+0.05, '%.02f' % b,  ha='center', va='bottom', fontsize=14)
for a, b in zip(N, Sn_QA_20):
    plt.text(a, b+0.002, '%.02f' % b,  ha='center', va='bottom', fontsize=14)
for a, b in zip(N, Sn_QA_100):
    plt.text(a, b+0.001, '%.02f' % b,  ha='center', va='bottom', fontsize=14)

font = {'family': 'DejaVu Sans',
        'weight': 'normal',
        'size': 15,
        }
plt.xlabel('Number of fibre span: N', font)
plt.ylabel('optimization value: |Sn|(ns)', font)
plt.legend(labels=['QA_100_iterations', 'QA_20_iterations',
                   'SA_100_iterations', 'SA_20_iterations'], loc='lower right', fontsize=11.5)
plt.grid()
plt.xscale('log')
plt.yscale('log')
plt.title('optimization performance: QA vs SA', font)
plt.show()
plt.savefig('value.png', dpi=200)
```

plot of optimization time

```python
#plt.figure(figsize=(4, 4))
plt.rcParams['figure.figsize'] = (8.0, 8.0)

plt1 = plt.plot(N, time_QA_100, marker='o', mec='blue', mfc='w')
plt2 = plt.plot(N, time_QA_20, marker='*', mec='red', mfc='w')
plt3 = plt.plot(N, time_SA_100, marker='v', mec='black', mfc='w')
plt4 = plt.plot(N, time_SA_20, marker='s', mec='brown', mfc='w')
for a, b in zip(N, time_SA_20):
    plt.text(a, b+0.05, '%.02f' % b, ha='center', va='bottom', fontsize=14)
for a, b in zip(N, time_SA_100):
    plt.text(a, b+0.05, '%.02f' % b,  ha='center', va='bottom', fontsize=14)
for a, b in zip(N, time_QA_20):
    plt.text(a, b+0.002, '%.02f' % b,  ha='center', va='bottom', fontsize=14)
for a, b in zip(N, time_QA_100):
    plt.text(a, b+0.001, '%.02f' % b,  ha='center', va='bottom', fontsize=14)

font = {'family': 'DejaVu Sans',
        'weight': 'normal',
        'size': 15,
        }
plt.xlabel('Number of fibre span: N', font)
plt.ylabel('optimization time(s)', font)
plt.legend(labels=['QA_100_iterations', 'QA_20_iterations',
                   'SA_100_iterations', 'SA_20_iterations'], loc='lower right', fontsize=11.5)
plt.grid()
plt.xscale('log')
plt.yscale('log')
plt.title('optimization time: QA vs SA', font)
plt.show()
plt.savefig('time.png', dpi=200)
```
